```python
import cv2
import numpy as np

img = cv2.imread('image/humenstanding0003.jpg')
mask = img[:, :, -1]

def padding(img):
    try:
        # 1. 그림의 x좌표, y좌표를 구한다
        y,x,c = img.shape
        print(f'y:{y}, x:{x}')
    except:
        print('파일을 확인 후 다시 시작하세요')
        raise
    
    # 2. 그림의 x좌표에 3/4를 곱하여 xt으로 정의한다.
    xt = round(x * (3/4))
    print('xt:', xt)
    # 3.     y값과 xt값을 비교한다 
    # 4-1    y값이 더 크다
    if y > x:
        # 5-1 y * 4/3을 하여 yt로 정의한다.
        yt = round(y * (4/3))
        print('yt:', yt)
        
        # 6-1 (yt - x)/2를 구하여 p로 정의한다.
        p = round((yt - x)/2)
        print('p:', p)
        
        #7-1 yt * y 크기의 검은 이미지를 만든다. (np.zeros(y, yt, 3)일듯)
        zero_pad = np.zeros([y,yt,3])
        print('img.shape:', img.shape)
        print('zero_pad.shape:', zero_pad.shape)
        
    
        # 자른 이미지를 소스 이미지에 붙인다.
#         zero_pad[0:, 0:] = img
        
#         cv2.imshow('result', zero_pad)
        
        #src_crop = zero_pad[0:y, p:p+x ]
        #cv2.imshow('a', src_crop)
        zero_pad[0:y, p:x+p, ] = img
#         src_crop
        cv2.imshow('result', zero_pad)
        

        
        
        
        
#         cv2.imshow('a', img)
#         cv2.imshow('b', zero_pad)       
        cv2.waitKey()
        cv2.destroyAllWindows()


        
        


        
        
        








# 4-2 xt값이 더 클 경우
# 5-2 (xt - y)/2를 구하여 p로 정의한다.
# 6-2 (0, x) * (0, p)를 0으로 패딩하고, (0, x) * (p, p + y)에 원본 이미지를 배치한다. 그리고 (0, x) * (p + y, xt)로 패딩한다.
padding(img)
```
