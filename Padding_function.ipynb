# 패딩 함수

```python
import cv2
import sys
import numpy as np


def padding(img):
    
    try:
        # 1. 그림의 x좌표, y좌표를 구한다
        y,x,c = img.shape
        print(f'y:{y}, x:{x}')
        
    except:
        print('파일을 확인 후 다시 시작하세요')
        raise
    
    # 2. 그림의 x좌표에 3/4를 곱하여 xt으로 정의한다.
    xt = round(x * (3/4))
    print('xt:', xt)
    
    # 3.     y값과 xt값을 비교한다 
    # 4-1    y값이 더 크다
    if y > x:
        # 5-1 y * 4/3을 하여 yt로 정의한다.
        yt = round(y * (4/3))
        print('yt:', yt)
        
        # 6-1 (yt - x)/2를 구하여 p로 정의한다.
        p = round((yt - x)/2)
        print('p:', p)
        
        #7-1 yt * y 크기의 검은 이미지를 만든다. (np.zeros(y, yt, 3)일듯)
        zero_pad = np.zeros([y,yt,3], dtype=np.uint8)
        print('img.shape:', img.shape)
        print('zero_pad.shape:', zero_pad.shape)
        
        zero_pad[0:y, p:x+p, ] = img
        
        return zero_pad
        
#         cv2.imshow('result', zero_pad)
#         cv2.waitKey()
#         cv2.destroyAllWindows()
    
    # x가 y보다 클 경우  
    if x > y:

        # 6-1 (xt - y)/2를 구하여 p로 정의한다.
        p = round((xt - y)/2)
        print('p:', p)
        
        #7-1 xt * x 크기의 검은 이미지를 만든다. (np.zeros(xt, x, 3)일듯)
        zero_pad = np.zeros([xt,x,3], dtype=np.uint8)
        print('img.shape:', img.shape)
        print('zero_pad.shape:', zero_pad.shape)
        
        zero_pad[p:y+p, 0:x, ] = img
        
        return zero_pad
        
#         cv2.imshow('result', zero_pad)    
#         cv2.waitKey()
#         cv2.destroyAllWindows()


        
```

# 패딩함수 적용 (결과보기)

패딩 전

```python
import cv2
img = cv2.imread('./image/img_ori1/g0.jpg')
cv2.imshow('img', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

패딩 후

```python
cv2.imshow('img',padding(img = cv2.imread('./image/img_ori1/g0.jpg')))  # 패딩함수 적용 후 이미지를 보여라

cv2.waitKey()
cv2.destroyAllWindows()
```

폴더 안에 있는 사진을 반복문을 통해 전체 패딩화

```python
import os

img_list_os = os.listdir('./image/img_ori1') # 경로설정하세요

for idx ,image in enumerate(img_list_os): # 디렉토리 파일 안에 있는 파일전체 for 문 통해 반복
    try:
        img = cv2.imread(f'./image/img_ori1/{image}') # 0번 째 부터 한 장씩 읽어와 img 변수에 저장
        jpg= padding(img) # 패딩 함수를 통한 아웃풋을 jpg 변수에 저장
        jpg = cv2.imwrite(f'./image/resize_with_padding1/{idx+1}.jpg', jpg) # 해당 디렉토리에 저장
    except:
        pass
    
    
    
    
    
    
    
    
```
