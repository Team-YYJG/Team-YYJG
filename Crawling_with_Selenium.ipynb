# 'selenium' 을 통해 구글크롤링 후 이미지 저장

```python
import os
import cv2
import sys
import time
import numpy as np
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 크롤링

path = './chromedriver'  # 경로
driver = webdriver.Chrome(path) # 크롬드라이버를 사용 할 수 있게 객체 생성 후 변수에 저장

# 주소를 넣으면 웹화면을 띄움 (원하는 주소 넣으면 됨)
driver.get('https://www.google.com/search?q=wide%20person%20with%20background&tbm=isch&hl=ko&tbs=rimg:CbgI3d4aKAHRYUcXH0vqv6TP8AEAsgIMCgIIABAAOgQIABAA&sa=X&ved=0CAIQrnZqFwoTCIDE_4TowvkCFQAAAAAdAAAAABAG&biw=912&bih=870') 

# _ 초 동안 대기
time.sleep(3)

# ()초 동안 대기, 스크롤 내릴 때
SCROLL_PAUSE_TIME = 3.5

# 스크롤 높이 가져오기, 자바스크립트 명령어를 사용
last_height = driver.execute_script("return document.body.scrollHeight")

# 구글은 스크롤을 아래로 내렸을 시 
#'검색결과 더보기' 버전이 2 가지라 try,except 로 2 가지 버전을 동시처리하게 함
try:
    while True:
        # 아래로 스크롤합니다
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # 페이지 로딩 대기
        time.sleep(SCROLL_PAUSE_TIME)

        # 새 스크롤 높이 계산 및 마지막 스크롤 높이와 비교
        new_height = driver.execute_script("return document.body.scrollHeight")

        # 새 높이와 이전 높이가 같다면
        if new_height == last_height:
            driver.find_element(By.CSS_SELECTOR,'.r0zKGf').click() # '검색결과 더보기' 란 클릭
            break
        last_height = new_height
        
except:
    while True:

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(SCROLL_PAUSE_TIME)

        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            try:
                driver.find_element(By.CSS_SELECTOR,'.mye4qd').click()        
            except:
                break
        last_height = new_height


images = driver.find_elements(By.CSS_SELECTOR,'.rg_i.Q4LuWd')   # '.rg_i.Q4LuWd' 클래스를 가진 이미지 전부 선택
img_url = [] # 이미지의 src 링크를 담을 빈리스트 생성 

for image in images: # 반복문
    try:
        image.click() # 이미지 한 장에 대하여 클릭 후
        imgurl = driver.find_element(By.CSS_SELECTOR, '.n3VNCb').get_attribute('src') # 이미지 한 장에 대하여 소스가져오기
        img_url.append(imgurl) # 이미지 한 장에 대하여 소스를 빈리스트에 업데이트

    except:
        pass
    

# 반복문을 통해 이미지 url 을 0 번 부터 돌아가며 이미지파일로 저장
for index, link in enumerate(img_url) :
    try:
        urllib.request.urlretrieve(link, f'./image/img_ori/p{0}{0}{index+1897}.jpg')     # urllib 사용하여 이미지링크를 이미지파일로 저장

    except:
        pass
        
```

# 패딩 함수 적용 후 폴더에 저장

패딩함수

```python
import os
import cv2
import sys
import time
import numpy as np
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By


# 패딩함수
def padding(img):
    
    try:
        # 1. 그림의 x좌표, y좌표를 구한다
        y,x,c = img.shape
        print(f'y:{y}, x:{x}')
        
    except:
        print('파일을 확인 후 다시 시작하세요')
        raise
    
    # 2. 그림의 x좌표에 3/4를 곱하여 xt으로 정의한다.
    xt = round(x * (3/4))
    print('xt:', xt)
    
    # 3.     y값과 xt값을 비교한다 
    # 4-1    y값이 더 크다
    if y > x:
        # 5-1 y * 4/3을 하여 yt로 정의한다.
        yt = round(y * (4/3))
        print('yt:', yt)
        
        # 6-1 (yt - x)/2를 구하여 p로 정의한다.
        p = round((yt - x)/2)
        print('p:', p)
        
        #7-1 yt * y 크기의 검은 이미지를 만든다. (np.zeros(y, yt, 3)일듯)
        zero_pad = np.zeros([y,yt,3], dtype=np.uint8)
        print('img.shape:', img.shape)
        print('zero_pad.shape:', zero_pad.shape)
        
        zero_pad[0:y, p:x+p, ] = img
        
        return zero_pad
        
#         cv2.imshow('result', zero_pad)
#         cv2.waitKey()
#         cv2.destroyAllWindows()
    
    # x가 y보다 클 경우  
    if x > y:

        # 6-1 (xt - y)/2를 구하여 p로 정의한다.
        p = round((xt - y)/2)
        print('p:', p)
        
        #7-1 xt * x 크기의 검은 이미지를 만든다. (np.zeros(xt, x, 3)일듯)
        zero_pad = np.zeros([xt,x,3], dtype=np.uint8)
        print('img.shape:', img.shape)
        print('zero_pad.shape:', zero_pad.shape)
        
        zero_pad[p:y+p, 0:x, ] = img
        
        return zero_pad
        
#         cv2.imshow('result', zero_pad)    
#         cv2.waitKey()
#         cv2.destroyAllWindows()


        
####################################################################################################
####################################################################################################
```

패딩 적용 후 폴더에 이미지파일 저장

```python
import cv2
import os
import sys
import numpy as np

img_list_os = os.listdir('./image/img_ori1')
img_list_os

for idx ,image in enumerate(img_list_os):
    try:
        img = cv2.imread(f'./image/img_ori1/{image}')
        jpg = padding(img)
        jpg = cv2.imwrite(f'./image/resize_with_padding1/qwe{idx}.jpg', jpg)
    except:
        pass
        
        
```
